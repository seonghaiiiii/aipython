```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
fig, axs = plt.subplots(2,2)
axs[0,0].plot(x,y)
axs[0,1].scatter(x,y)
axs[1,0].bar(x,y)
axs[1,1].barh(x,y)
fig.suptitle('Drawing Graph')
plt.show()
```

```python

sns.scatterplot(x='petal_length',y='petal_width'
,hue='species',data=iris)
plt.title('Scatter Plot by seaborn', fontsize=20)
plt.show()
```

```python
sns.countplot(x='species',data=iris)
plt.show()
```

```python
sns.barplot(x='species',y='petal_length',data=iris, errorbar='sd')
plt.show()
```

```python
sns.barplot(x='species',y='petal_width',data=iris, errorbar='sd')
plt.show()
```

```python
sns.pairplot(iris,hue='species')
plt.show()
```

```python
sns.regplot(x='petal_length',y='petal_width', data=iris)
plt.show()
```

```python
df=iris.corr()
sns.heatmap(data=df,annot=True,fmt='.2f',cbar=False)
plt.show()
```

```python
sns.barplot(x='species',y='body_mass_g',hue='sex',data=pg, errorbar='sd')
plt.show()
```
