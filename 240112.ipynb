```python
#꽃받침 길이 시각화
plt.rc('font', family='NanumGothic')
ir = iris.head(10)
plt.bar(ir.index,ir['sepal_length'])
plt.title('꽃받침 길이 시각화')
plt.xlabel('Data')
plt.ylabel('꽃받침의 길이')
```

```python
#붓꽃 종류별 꽃 받침 길이의 평균에 대한 시각화
plt.rc('font', family='NanumGothic')
plt.bar(iris['species'],iris['sepal_length'])
plt.title('품종별 sepal length 그래프')
plt.xlabel('종류')
plt.ylabel('꽃받침 길이의 평균값')
```

```python
plt.rc('font', family='NanumGothic')
plt.bar(['setosa','versicolor','virginica'], iris.groupby('species')['sepal_length'].mean())

# 그래프 제목과 라벨 설정
plt.title('품종별 Sepal Length의 평균값')
plt.xlabel('종류')
plt.ylabel('꽃받침 길이의 평균값')

# 그래프 표시
plt.show()
```

```python
#각 특징 값들의 빈도수 그래프
plt.rc('font', family='NanumGothic')
plt.hist(iris['sepal_length'],bins=3,label='sepal_length')
plt.hist(iris['sepal_width'],bins=3,label='sepal_width')
plt.hist(iris['petal_length'],bins=3,label='petal_length')
plt.hist(iris['petal_width'],bins=3,label='petal_length')
plt.title('각 속성 값들의 빈도수 히스토그램')
plt.xlabel('Value')
plt.ylabel('빈도수')
plt.legend()
plt.show()
```

```python
#산점도
plt.rc('font', family='NanumGothic')
plt.scatter(iris['sepal_length'],iris['petal_length'])
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
```

#타이타닉 데이터

```python
#타이타닉 파일 읽어오기
tt.head()
```

```python
#기본 정보 출력
tt.info()
```

```python
#기초 통계량 출력
tt.describe()
```

```python
#요금 기준 내림차순 정렬
tt.sort_values(by=['fare'],ascending=False,inplace=True)
tt
```

```python
#생존자별 인원수
tt['survived'].value_counts()
```

```python
#결측치 확인
tt.isnull().sum()
```

```python
#결측치 컬럼 (deck) 삭제
# tt.drop('deck',axis=1,inplace=True)
tt.columns
```

```python
#탑승 항구 최빈값 확인
tt['embarked'].value_counts()
```

```python
#탑승 항구 결측치 최빈값 대체
tt.fillna({'embarked':'S'},inplace=True)
tt.isnull().sum()
```

```python
#나이 결측치 평균값 대체
tt.fillna({'age':tt['age'].mean()},inplace=True)
tt.isnull().sum()
```

```python
#항구별 pclass 컬럼의 평균값 출력
tt.groupby('embarked')['pclass'].mean()
```

```python
#pclass,sex 컬럼의 평균값 출력
tt.groupby(['pclass','sex']).mean()
```

```python
#좌석 등급별 생존자 확인 (시본)
sns.countplot(x='pclass',hue='survived',data=tt)
plt.xlabel('pclass')
plt.ylabel('count')
```

```python
#상관관계 확인(시본)
a = tt.corr()
sns.heatmap(data=a,annot=True,cmap='Blues')
```
